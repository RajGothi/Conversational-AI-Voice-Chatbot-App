```python
from pydub import AudioSegment

def convert_audio(input_file, output_file):
    # Load the audio file
    audio = AudioSegment.from_file(input_file)
    
    # Convert to mono (1 channel)
    audio = audio.set_channels(1)
    
    # Convert to 16-bit samples
    audio = audio.set_sample_width(2)
    
    # Convert to 8000 Hz
    audio = audio.set_frame_rate(8000)
    
    # Export the converted audio
    audio.export(output_file, format="wav")
```

```python
input_audio_file = 'Recording.wav'
output_audio_file = 'format_recording.wav'
convert_audio(input_audio_file, output_audio_file)
```

```python
import base64
import wave
import webrtcvad

# Function to read audio file
def read_audio_file(file_path):
    with open(file_path, 'rb') as f:
        audio_data = f.read()
    return audio_data

# Function to apply VAD
def apply_vad(audio_data):
    vad = webrtcvad.Vad()
    vad.set_mode(1)  # Set aggressive mode (0-3)

    frames = []
    chunk_size = 320  # 20ms for 8000Hz sample rate

    # Process audio in chunks
    for i in range(0, len(audio_data), chunk_size):
        frame = audio_data[i:i + chunk_size]
        if len(frame) == chunk_size:
            is_speech = vad.is_speech(frame, sample_rate=8000)
            frames.append((frame, is_speech))

    return frames

# Example usage
if __name__ == '__main__':
    audio_file_path = 'format_recording.wav'

    # Step 1: Read audio file
    audio_data = read_audio_file(audio_file_path)

    # Step 2: Encode audio data as base64
    encoded_payload = base64.b64encode(audio_data).decode('utf-8')

    # Step 3: Decode base64 data
    decoded_audio_data = base64.b64decode(encoded_payload.encode('utf-8'))

    # Step 4: Apply VAD
    detected_frames = apply_vad(decoded_audio_data)

    # Process detected frames as needed
    for frame, is_speech in detected_frames:
        if is_speech:
            print("Speech detected in this frame.")
        else:
            print("Non-speech frame.")

    # Optionally, you can re-encode the frames to base64 for storage or transmission
```
